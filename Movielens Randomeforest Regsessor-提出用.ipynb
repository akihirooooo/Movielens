```python
import pandas as pd
import numpy as np
```

```python
#行列に名前付け
data_cols = ['user_id', 'movie_id', 'rating', 'timestamp']
item_cols = ['movie_id', 'movie_title', 'release_date', 'video_release_date', 'IMDb_URL', 'unknown',
             'Action', 'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary', 'Drama',
             'Fantasy', 'Film_Noir', 'Horror', 'Musical', 'Mystery', 'Romance ', 'Sci_Fi', 'Thriller', 'War', 'Western']
user_cols = ['user_id', 'age','gender', 'occupation', 'zip_code']
```

```python
#データ読み込み
users = pd.read_csv('ml-100k/u.user', sep='|', names = user_cols, encoding = 'latin-1')
item = pd.read_csv('ml-100k/u.item', sep='|', names = item_cols, encoding = 'latin-1')
data = pd.read_csv('ml-100k/u.data', sep='\t',names = data_cols, encoding = 'latin-1')
```

```python
#user_idが行、movie_idが列になるようlong形式からwide形式へ変換
data1 = data.pivot(index = "user_id", columns = "movie_id", values = "rating")
data1.head()
```

```python
#インデックスを振りなおす
data2 = data1.reset_index()
```

```python
data2.head()
```

```python
#NaNを0に変換
data3 = data2.fillna(0)
```

```python
#目的変数と説明変数のuser_idを統一するためにleft joinでつなげる
user_data = pd.merge(users2, data3, on = "user_id", how = "left")
```

```python
user_data.head()
```

```python
#学習用とテスト用に分けるモジュールのインポート
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(user_data, test_size = 0.2, random_state = 123)
```

```python
print(train.shape)
print(test.shape)
```

```python
train.head()
```

```python
#目的変数の作成
train_y = train.iloc[:,820:]
test_y = test.iloc[:,820:]
```

```python
#説明変数の作成（user_idは除き、movie_idの手前まで）
train_x = train.iloc[:,1:820]
test_x = test.iloc[:,1:820]
```

```python
test_x.head()
```

```python
#ランダムフォレストのインポート
from sklearn.ensemble import RandomForestRegressor
```

```python
forest = RandomForestRegressor(n_estimators = 100, random_state = 123, n_jobs = -1)
```

```python
#学習用データセットで学習させる
forest.fit(train_x, train_y)
```

```python
#テスト用データセットの説明変数を入れ、目的変数をpredictionとして出す
prediction = forest.predict(test_x)
```
